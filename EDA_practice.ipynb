```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('us-presidential-election-county-results-2004-through-2012.csv')
df=df.sort_values(by=['pct_dem'])
df.head()
```

```python
df=df.drop(df.index[[519,3673]])
df
```

```python
df.shape
```

```python
df.describe()
```

```python
states_2004=df[df.year==2004]
states_2008=df[df.year==2008]
states_2012=df[df.year==2012]
```

```python
states_2004.head()
```

```python
states_2008.head()
```

```python
states_2012.head()
```

```python
all_states=df['state'].unique()
all_states
```

```python
dem_states_2004=[]
rep_states_2004=[]
```

```python
for i in all_states:
    dem_states_2004.append(sum(states_2004[states_2004.state==i]['vote_dem']))
    rep_states_2004.append(sum(states_2004[states_2004.state==i]['vote_rep']))
```

```python
votes_2004=pd.DataFrame(np.c_[dem_states_2004,rep_states_2004],index=all_states)
votes_2004.columns=['Democrat','Republic']
votes_2004.plot(kind='bar',figsize=(25,25))
plt.show()
```

```python
dem_states_2008=[]
rep_states_2008=[]

for i in all_states:
    dem_states_2008.append(sum(states_2008[states_2008.state==i]['vote_dem']))
    rep_states_2008.append(sum(states_2008[states_2008.state==i]['vote_rep']))

votes_2008=pd.DataFrame(np.c_[dem_states_2008,rep_states_2008],index=all_states)
votes_2008.columns=['Democrat','Republic']
votes_2008.plot(kind='bar',figsize=(25,25))
plt.show()
```

```python
dem_states_2012=[]
rep_states_2012=[]

for i in all_states:
    dem_states_2012.append(sum(states_2012[states_2012.state==i]['vote_dem']))
    rep_states_2012.append(sum(states_2012[states_2012.state==i]['vote_rep']))

votes_2012=pd.DataFrame(np.c_[dem_states_2012,rep_states_2012],index=all_states)
votes_2012.columns=['Democrat','Republic']
votes_2012.plot(kind='bar',figsize=(25,25))
plt.show()
```

```python
swing_state=df[df.state=='OH']
swing_state=swing_state.append(df[df.state=='NV'])
swing_state=swing_state.append(df[df.state=='IA'])
swing_state=swing_state.append(df[df.state=='FL'])
swing_state=swing_state.append(df[df.state=='IN'])
swing_state=swing_state.append(df[df.state=='MN'])

swing_state
```

```python
swing_states=swing_state['state'].unique()
swing_states
```

```python
margins_avg=[]
for i in swing_states:
    avg=sum(swing_state[swing_state.state==i]['margin_winner_over_runnerup'])/len(swing_state[swing_state.state==i]['margin_winner_over_runnerup'])
    margins_avg.append(avg)


margins_avg
```

```python
margin_frame=pd.DataFrame(np.c_[margins_avg],index=swing_states)
margin_frame.columns=['Average margins']
margin_frame.plot(kind='bar',figsize=(20,20))
plt.show()
```
